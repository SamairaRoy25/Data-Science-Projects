### Setup & Data Cleaning

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

sns.set_theme(style="whitegrid")
```

```python
df = pd.read_csv("/kaggle/input/select-world-bank-commodity-price-data/commodity_prices.csv", index_col="date", parse_dates=True)
df.head()
```

```python
df.drop("Unnamed: 0", axis=1, inplace=True)
df.isna().sum()
```

### Analysis

#### Oil

```python
plt.figure(figsize=(15, 5))
oil_plot = sns.lineplot(data=df[["oil_brent", "oil_dubai"]].rolling(20).mean())
plt.title("Oil Prices")
plt.ylabel("Price per Barrel (USD)")
plt.xlabel("Date")
h, _ = oil_plot.get_legend_handles_labels()
oil_plot.legend(h, ["Brent", "Dubai"], loc="upper left")
```

#### Coffee

```python
plt.figure(figsize=(15, 5))
coffee_plot = sns.lineplot(data=df[["coffee_arabica", "coffee_robustas"]].rolling(20).mean())
plt.title("Coffee Prices")
plt.ylabel("Price per Kilogram (USD)")
plt.xlabel("Date")
h, _ = coffee_plot.get_legend_handles_labels()
coffee_plot.legend(h, ["Arabica", "Robustas"], loc="upper left")
```

#### Tea

```python
plt.figure(figsize=(15, 5))
tea_plot = sns.lineplot(data=df[["tea_columbo", "tea_kolkata", "tea_mombasa"]].rolling(20).mean())
plt.title("Tea Prices")
plt.ylabel("Price per Kilogram (USD)")
plt.xlabel("Date")
h, _ = tea_plot.get_legend_handles_labels()
tea_plot.legend(h, ["Columbo", "Kolkata", "Mombasa"], loc="upper left")
```

#### Sugar

```python
plt.figure(figsize=(15, 5))
sugar_plot = sns.lineplot(data=df[["sugar_eu", "sugar_us", "sugar_world"]].rolling(20).mean())
plt.title("Sugar Prices")
plt.ylabel("Price per Kilogram (USD)")
plt.xlabel("Date")
h, _ = sugar_plot.get_legend_handles_labels()
sugar_plot.legend(h, ["European Union", "United States", "World"], loc="upper left")
```
